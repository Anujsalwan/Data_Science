14.	Which shark has the highest return on investment (ROI) based on deal_amount vs. deal_equity?

```python
shark_deals = shark[
    (shark['deal_valuation'] > 0) &
    (shark['amount_per_shark'] > 0) &
    (shark['equity_per_shark'] > 0)
].copy()


shark_deals['value_acquired'] = shark_deals['deal_valuation'] * (shark_deals['equity_per_shark'] / 100)


sharks = {
    'Ashneer Grover': 'ashneer_deal',
    'Anupam Mittal': 'anupam_deal',
    'Aman Gupta': 'aman_deal',
    'Namita Thapar': 'namita_deal',
    'Vineeta Singh': 'vineeta_deal',
    'Peyush Bansal': 'peyush_deal',
    'Ghazal Alagh': 'ghazal_deal'
}

roi_data = []
for shark_name, deal_col in sharks.items():

    shark_investments = shark_deals[shark_deals[deal_col] == 1]


    total_value_acquired = shark_investments['value_acquired'].sum()


    total_amount_invested = shark_investments['amount_per_shark'].sum()


    if total_amount_invested > 0:
        roi = total_value_acquired / total_amount_invested
    else:
        roi = 0.0

    roi_data.append({
        'Shark': shark_name,
        'Total Amount Invested (Lakhs)': total_amount_invested,
        'Total Value Acquired (Lakhs)': total_value_acquired,
        'ROI (Value/Investment Ratio)': roi
    })


roi_df = pd.DataFrame(roi_data)
roi_df_sorted = roi_df.sort_values(by='ROI (Value/Investment Ratio)', ascending=False)
roi_df_sorted['ROI (Value/Investment Ratio)'] = roi_df_sorted['ROI (Value/Investment Ratio)'].round(2)
roi_df_sorted['Total Amount Invested (Lakhs)'] = roi_df_sorted['Total Amount Invested (Lakhs)'].round(2)
roi_df_sorted['Total Value Acquired (Lakhs)'] = roi_df_sorted['Total Value Acquired (Lakhs)'].round(2)

print(roi_df_sorted.to_markdown(index=False))
```

16.	What is the relationship between pitcher_ask_amount and deal_amount? Do pitchers who ask for less tend to secure more deals?

```python
correlation = shark['pitcher_ask_amount'].corr(shark['deal_amount'])
plt.figure(figsize=(10, 6))
sns.scatterplot(x='pitcher_ask_amount', y='deal_amount', data=shark)
plt.title(f'Scatter Plot: Pitcher Ask Amount vs. Deal Amount (Correlation: {correlation:.2f})', fontsize=14)
plt.xlabel('Pitcher Ask Amount (in Lakhs)', fontsize=12)
plt.ylabel('Deal Amount (in Lakhs)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

```

1. Relationship between pitcher_ask_amount and deal_amount

The Pearson correlation coefficient between the pitcher_ask_amount and deal_amount is $\mathbf{-0.08}$.

This value is very close to zero, indicating an extremely weak, almost non-existent linear relationship between the amount a pitcher asks for and the final amount secured in the deal.

The Scatter Plot visually confirms this, showing the data points are widely scattered with no clear trend.

17.	Identify if there is a pattern in episode-wise deal closures—do certain episodes see more investments?

```python
deals_df = shark[shark['deal_amount'] > 0]


deals_per_episode = deals_df.groupby('episode_number').size().reset_index(name='Number_of_Deals')


deals_per_episode = deals_per_episode.sort_values(by='episode_number')


plt.figure(figsize=(12, 6))
sns.barplot(x='episode_number', y='Number_of_Deals', data=deals_per_episode)
plt.title('Number of Deal Closures Per Episode', fontsize=16)
plt.xlabel('Episode Number', fontsize=12)
plt.ylabel('Number of Deals', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

Yes, there is a pattern in episode-wise deal closures, primarily characterized by consistency but with a few episodes securing significantly more investment.The most common outcome is $\mathbf{1}$ or $\mathbf{2}$ deals per episode, but certain episodes clearly see more investments.

18.	Analyze the impact of equity dilution on deal_valuation. Are lower equity asks leading to higher deal valuations?

```python
deals_df = shark[shark['deal_valuation'] > 0].copy()
correlation = deals_df['ask_equity'].corr(deals_df['deal_valuation'])
plt.figure(figsize=(10, 6))
sns.scatterplot(x='ask_equity', y='deal_valuation', data=deals_df)
plt.title(f'Equity Ask vs. Deal Valuation (Successful Deals) \nPearson Correlation: {correlation:.2f}', fontsize=14)
plt.xlabel('Pitcher Ask Equity (%)', fontsize=12)
plt.ylabel('Deal Valuation (in Lakhs)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.yscale('log')
plt.ylabel('Deal Valuation (in Lakhs, Log Scale)', fontsize=12)
plt.show()
```

Analysis of Equity Ask vs. Deal Valuation
1. CorrelationThe Pearson correlation coefficient between ask_equity and deal_valuation (for successful deals) is $\mathbf{-0.52}$.
 
This indicates a moderate negative correlation.In practical terms, this supports the hypothesis: lower equity asks are generally associated with higher deal valuations. Pitchers who are willing to give up less equity tend to be valued more highly by the sharks.

This is a common pattern in investing, where high-growth companies with strong negotiation power seek to minimize dilution.

2. Visual Inspection (Scatter Plot)The scatter plot below, using a logarithmic scale for Deal Valuation to handle the large spread in values, visually confirms the negative trend.
  
As the percentage of ask_equity increases on the horizontal axis, the data points for deal_valuation tend to cluster at lower values.
