Q6. Explain the concept of stacking in numpy with examples.
How is it different from concatenation. Give any 3 differences between them.

### Stacking in Numpy
Stacking is a method of joining multiple arrays along a new axis. The arrays must have the same shape along the non-stacking axes. The stacking operation effectively increases the dimensionality of the resulting array.

1. np.vstack() (Vertical Stacking): Stacks arrays vertically (row-wise). The arrays must have the same number of columns.
2. np.hstack() (Horizontal Stacking): Stacks arrays horizontally (column-wise). The arrays must have the same number of rows.
3. np.dstack() (Depth Stacking): Stacks arrays along the third axis (depth). This is useful for creating 3D arrays from 2D arrays. The arrays must have the same shape.


#### Differences between Stacking and Concatenation
Concatenation is a more general operation that joins arrays along an existing axis. It's performed using the np.concatenate() function. The main difference lies in how they combine arrays and the resulting array's shape.

Here are three key differences between stacking vs concatenation:

Axis of Joining : Joins arrays along a new axis. It increases the dimensionality of the resulting array vs Joins arrays along an existing axis. It maintains the original dimensionality.

Array Shapes : Requires arrays to have the same shape along the non-stacking axes. For vstack, the number of columns must be the same. For hstack, the number of rows must be the same vs Requires arrays to have the same shape along all axes except the one being concatenated.

Functionality : Specialized functions for specific stacking directions: vstack (vertical), hstack (horizontal), dstack (depth) vs A single, general-purpose function (np.concatenate) that requires an axis argument to specify the joining direction.
