```python
def custom_q11(shape,**dimen):
    area=0
    try:
        if shape.lower()=="circle":
            area=dimen['r']*3.14
        elif shape.lower()=="square":
            area=dimen['s']**2
        elif shape.lower()=="rectangle":
            area=dimen['l']*dimen['w']
        else:
            print("Enter the right shape")
    except:
        if shape.lower()=="circle":
            print("Enter the radius of Circle")
        elif shape.lower()=="square":
            print("Enter the Side of Square")
        elif shape.lower()=="rectangle":
            print("Enter the length and breadth of rectangle")
        
            
    return area
print(custom_q11(shape="CIRCLe",r=10))
print(custom_q11(shape="sQuare",s=10))
print(custom_q11(shape="rectangle",l=10,w=11))
```

```python
def reconcile_lists(l1, l2):

  matched = list(set(l1) & set(l2))
  only_in_l1 = list(set(l1) - set(l2))
  only_in_l2 = list(set(l2) - set(l1))

  return {
      'Matched': matched,
      'Only in l1': only_in_l1,
      'Only in l2': only_in_l2
  }


l1 = ['January', 'February', 'March', 'May', 'June', 'September', 'December']
l2 = ['January', 'February', 'April', 'June', 'October', 'December']
reconcile_lists(l1, l2)
```

```python
def prime_num(n):
    count=0
    for i in range(1,n+1):
        if n%i==0:
            count+=1
    if count==2:
        return "Prime number"
    else:
        return "Not A Prime number"
print(prime_num(9))
print(prime_num(3))
```
