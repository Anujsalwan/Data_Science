3.	Explain the concept of missing values. How can you identify the missing values in a Pandas DataFrame ?
What are the different ways of treating/Imputing/Deleting the missing values.
Explain with example.


##### Missing values occur in datasets when some entries for certain variables are not recorded or are unavailable.
In Pandas, missing values are usually represented by:NaN (Not a Number) for numeric columns, None or NaT for object/datetime columns

### different ways of treating/Imputing/Deleting the missing values:

#### 1. Deletion
#### Remove rows or columns containing missing values.

#Drop rows with any missing value:
 df_dropped_rows = df.dropna()

#Drop columns with any missing value:
 df_dropped_cols = df.dropna(axis=1)

#### 2. Imputation (Filling)
#### Replace missing values with a substitute.

##### a) Using a constant value: df_filled_const = df.fillna(0)  # replace NaN with 0
##### b) Using statistical measures :df["Age"].fillna(df["Age"].mean(), inplace=True)   # mean
                        df["Age"].fillna(df["Age"].median(), inplace=True) # median
                        df["Name"].fillna(df["Name"].mode()[0], inplace=True) # mode


#### 3. Forward fill / Backward fill
##### df_ffill = df.fillna(method="ffill")  # fill using previous value
df_bfill = df.fillna(method="bfill")  # fill using next value



## Examples

```python
#creating a dummy dataset containing missing values
import numpy as np


df = pd.DataFrame({
    "Name": ["Anuj", "Amit", None, "Lalit", "Arun"],
    "Age": [25, np.nan, 22, np.nan, 28],
    "Score": [85, 90, np.nan, 75, None]
})

print("Original DataFrame with Missing Values:\n", df)
```

```python
#identifying missing values
print(df.isnull())
```

```python
#deleting missing values
df.dropna()
```

```python
#Fiiling missing values with a constant
df.fillna(0)
```

```python
#Forward Fill
df.ffill()
```

```python
#Backward Fill
df.bfill()
```
