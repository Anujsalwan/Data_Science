### Step 1: Basic Structure

```python
def welcome():
    print("Welcome to Library Management System.")
```

```python
library = {
    "Python Basics": {"author": "Arihant", "available": True,"issued":""},
    "Data Science": {"author": "Anuj", "available": False,"issued":"Lalit"}
}
```

```python
def display():
    for i,j in library.items():
        print("The Book name is",i,"and Author is",j.get("author"))
```

## Step 2: Adding Books

```python
def add_book(name,author):
    if (name not in [i for i in library.keys()]) and (author not in [i["author"] for i in library.values()]):
        library[name]={"author": author, "available": True}
        print("Book added Sucessfully")
    else:
        print("Book with Same Name or Author Exists")
```

## Step 3: Issuing Books

```python
def issue_book(name,issuer):

    def display():
        try:
            if library[name]["available"]==True:
                print("Books is available")
            else:
                print("Books is Not Available")
        except:
            print("No Book with This name exist in the Liabrary")
    try:
        if library[name]["available"]==True:
            library[name]["available"]=False
            library[name]["issued"]=issuer
            print("Book",name,"Has Been Issued to",issuer)    
        
        else:
            print("The book has been issued by",library[name]["issued"])
            
    except:
        print("No Book with This name exist in the Liabrary")
```

### Step 4: Returning Books

```python
def return_book(name):
    try:
        if (library[name]["available"]==False) and (name in [i for i in library.keys()]):
            library[name]["available"]=True
            print("Book has been returned Sucessfully")
        else:
            print("Books Not Issued")
    except KeyError:
        print("No Book avaialble with this name")
    except Exception as e:
        print("Unexpcted error :", e)
```

## Step 5: Login System

```python
credentials = {"admin": "admin123", "librarian": "lib123"}
def login_system():
    for i in range(3):
        
        id1=input("Enter the id of users/press q to quit")
        if id1.lower()=="q":
            break       
        passw=input("Enter the password/press q to quit")
        if passw.lower()=="q":
            break
        
        if (id1 in credentials) and (credentials.get(id1)==passw):
            print("Correct ID & Password Successful Login")
            return "Right login id & pass"
            break
        else:
            print("Wrong ID & Password Please try Again")              
```

## Step 6: View Issued Books

```python
def view_issued_books():
    issued_books_found = False
    for i,j in library.items():
        if j.get("available")==False:
            print("The Book's name is", i, "& issued to", j.get("issued"))
            issued_books_found = True
    if not issued_books_found:  
        print("No books has been issued (All Books are available for Issues)")
```

## Step 7: Main Menu

```python
count=0
while count<1:
            welcome()
            a=login_system()
            count+=1
             
            if a=="Right login id & pass":
                while True:
                    selection=input(""" Press q/Q to quit or Select the following options
                                    1. View Options
                                    2. Add a Book
                                    3. Issue a Book
                                    4. Return a Books
                                    5. View Issued Books               
                                    """)
            
                    if selection.lower()=="q":
                        print("You have exist succesfully")
                        break
                    elif selection=="1":
                        display()
                    elif selection=="2":
                        name=input("Enter the Name of Book to be added")
                        author=input("Enter the author of Book of added")
                        add_book(name,author)
                    elif selection=="3":
                        name1=input("Enter the name of Book to be issued")
                        issuer1=input("Enter the name of person to whome book to be issed")
                        issue_book(name1,issuer1)
                    elif selection=="4":
                        book=input("Name of Book to be returned")
                        return_book(book)
                    elif selection=="5":
                        view_issued_books()
                    else:
                        print("Select the correction Option from 1 to 5 or Q/q to quit")
```
