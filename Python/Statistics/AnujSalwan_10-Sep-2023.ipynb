```python
df['Outliners_Total_salary']=(df.Total_salary-df.Total_salary.mean())/df.Total_salary.std()
```

```python
df[df['Outliners_Total_salary']<-3]
```

```python
def outliners_treatment(col):
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range=Q1-(1.5*IQR)
    upper_range=Q3+(1.5*IQR)
    return lower_range,upper_range
```

```python
lr,ur=outliners_treatment(df['Total_salary'])
```

```python
df['Total_salary']=np.where(df['Total_salary']<lr,lr,df['Total_salary'])
df['Total_salary']=np.where(df['Total_salary']>ur,ur,df['Total_salary'])
sns.boxplot(data=df,x='Total_salary');
```

```python
df.drop(axis=0,columns='Outliners_Total_salary',inplace=True)
```

```python
#Question 2 2. What is the likelihood of a salaried person buying a Sedan?
number_of_salaries_sedans = df[ (df['Make'] == 'Sedan') & (df['Profession'] == 'Salaried') ]
number_of_sedans=df[ (df['Make'] == 'Sedan')]
percentage=(len(number_of_salaries_sedans)/len(number_of_sedans))*100
print('likelihood of a salaried person buying a Sedan',round(percentage,2))
```
